```python
from math import sqrt
import copy

class Point:
    """A class represents a point in 2D space"""
    
class Rectangle:
    """A class represents a rectangle
       Attributes: lower left corner, width, height 
    """    

class Circle:
    """A class represents a circle
       Attributes: center, radius
    """
```

```python
def print_point(point):
    print('(%g, %g)'%(point.x, point.y))

def distance_between_points(point1,point2):
    d = sqrt((point1.x-point2.x)**2 + (point1.y - point2.y)**2)
    return d

def find_center(rectangle):
    point = Point()
    point.x = rectangle.corner.x + rectangle.width/2
    point.y = rectangle.corner.y + rectangle.height/2
    return point

def change_rectangle(rectangle, dwidth, dheight):
#     rectangle = Rectangle()
    rectangle.width += dwidth
    rectangle.height += dheight
    
def move_rectangle(rectangle, dx, dy):
    rectangle.corner.x += dx
    rectangle.corner.y += dy

def move_copied_rectangle(rectangle, dx, dy):
    copied_rec = copy.deepcopy(rectangle)
    move_rectangle(copied_rec,dx,dy)
    return copied_rec

def point_in_circle(circle, point):
    """To test if a point is in a circle
       True if the point lies in or on the boundary of the circle or on the circle, false otherwise
    """
    d = distance_between_points(circle.center, point)
    return d <= circle.radius

def my_rect_in_circle(circle, rec):
    rec.upper_right_corner = Point()
    rec.upper_right_corner.x = rec.corner.x + rec.width
    rec.upper_right_corner.y = rec.corner.y + rec.height

    d1 = distance_between_points(circle.center, rec.corner)
    d2 = distance_between_points(circle.center, rec.upper_right_corner)
    
    return d1 <= circle.radius and d2 <= circle.radius

def rect_in_circle(circle,rect):
    """Checks whether the corners of a rect fall in/on a circle.

    rect: Rectangle object
    circle: Circle object
    """
    p = copy.copy(rect.corner)
    print_point(p)
    if not point_in_circle(circle,p):
        return False

    p.x += rect.width
    print_point(p)
    if not point_in_circle(circle,p):
        return False

    p.y -= rect.height
    print_point(p)
    if not point_in_circle(circle,p):
        return False

    p.x -= rect.width
    print_point(p)
    if not point_in_circle(circle,p):
        return False

    return True
```

```python
circle1 = Circle()
circle1.center = Point()
circle1.center.x = 0
circle1.center.y = 0
circle1.radius = 10

rec1 = Rectangle()
rec1.corner = Point()
rec1.corner.x = -1
rec1.corner.y = -1
rec1.width = 2
rec1.height = 2

point1 = Point()
point1.x = 0
point1.y = 0

a = my_rect_in_circle(circle1, rec1)
print(a)

b = point_in_circle(circle1, point1)
print(b)

c= rect_in_circle(circle1,rec1)
print(c)
```
